```python
import numpy as np
def splite_train_test(data,test_ratio):
    np.rendom.seed(42)
    shuffled = np.random.permutation(len(data))
    print(shuffled)
    test_set_size = int(len(data) * test_ratio)
    test_indices = shuffled[:test_set_size]
    train_indices = shuffled[test_set_size]
    return data.iloc[train_indices], data.iloc[test_indices]
```

```python
import matplotlib.pyplot as plt
```

```python
# Line Diagram
housing.area.plot(kind = 'line', color = 'g', label = 'area', linewidth = 1, alpha = 0.5, grid = True, linestyle = ':')
housing.price.plot(color = 'r', label = 'price', linewidth = 1, alpha = 0.5, grid = True, linestyle = '-.')
plt.legend(loc='upper right')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Line Plot')
plt.show()
```

```python
# Scatter Plot
# x = area, y = price
housing.plot(kind = 'scatter', x = 'area', y = 'price', alpha = 0.5, color = 'red')
plt.xlabel('Area')
plt.ylabel('Price')
plt.title('Area Price Scatter Plot')
```

```python
# Histogram
# bins = number of bar in figure
housing.price.plot(kind = 'hist', bins = 50, figsize = (12,12))
plt.show()
```
